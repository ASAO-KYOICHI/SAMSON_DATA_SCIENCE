```python
import sys
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# このnotebookの存在するディレクトリのパス（必要に応じて変更！）
path = "/root/userspace/contents/materials_homework/"
gzfilename = path + 'test_img_4.png'

# 画像を読み込む。
gray = cv2.imread(gzfilename, cv2.IMREAD_GRAYSCALE)#グレースケールで読み出し
img  = cv2.imread(gzfilename)#そのままRGBで読み出し

binary = cv2.inRange(gray, 0.01, 55)#グレースケール（2色）表示

plt.figure(figsize=(8.73/1.5, 4.66/1.5), dpi = 100)#表示する画面サイズ
plt.imshow(binary)#グレースケール化した画像を表示する
plt.show()

# 輪郭抽出
contours = cv2.findContours(binary,
                            cv2.RETR_EXTERNAL,
                            cv2.CHAIN_APPROX_TC89_L1)[0]

# エリア長が一定以上の部分のみ残す。
area_thresh = 40
contours = list(filter(lambda x: cv2.arcLength(x,True) > area_thresh, contours))

g_list = []#空のリストを作成
g_re_list = []#空のリストを作成
counter = 0#カウンタを定義
x_jiku = []#空のリスト
y_jiku = []#空のリスト

# 輪郭を矩形で囲む。
for cnt in contours:
    # 輪郭に外接する長方形を取得する。
    x, y, width, height = cv2.boundingRect(cnt)#一定以上の面積がある箇所の原点と縦・横長を出す
    # 描画する。
    if (width > 20) & (height > 20) & (width < 100) & (height < 100):
        #エリアの大きさが一般的な文字サイズ位である場合
        x_jiku.append(x)#X軸情報を加算
        y_jiku.append(y)#Y軸情報を加算
        cv2.rectangle(img,#囲いを書き込む画像データ
                      (x, y),#バウンディングボックスの原点
                      (x + width, y + height),#原点からの対角点
                      color=(0, 255, 0),#色：green
                      thickness = 2)#描写線の厚み
        g_list.append( binary[y:y+height, x:x+width] )#リストに追加
        g_once = np.zeros((22,22))
        g_once = cv2.resize(binary[y:y+height, x:x+width], (22, 22))#, interpolation=cv2.INTER_NEAREST)
        #空欄を挿入する処理↓
        g_once = np.insert(g_once, 22, 0, axis = 0)
        g_once = np.insert(g_once, 22, 0, axis = 0)
        g_once = np.insert(g_once, 22, 0, axis = 0)
        g_once = np.insert(g_once, 22, 0, axis = 1)
        g_once = np.insert(g_once, 22, 0, axis = 1)
        g_once = np.insert(g_once, 22, 0, axis = 1)
        g_once = np.insert(g_once, 0, 0, axis = 0)
        g_once = np.insert(g_once, 0, 0, axis = 0)
        g_once = np.insert(g_once, 0, 0, axis = 0)
        g_once = np.insert(g_once, 0, 0, axis = 1)
        g_once = np.insert(g_once, 0, 0, axis = 1)
        g_once = np.insert(g_once, 0, 0, axis = 1)
        g_re_list.append( g_once )
        #リサイズ（28*28）してリストに追加
        counter += 1#カウンタをインクリメント

cv2.imwrite(path + "ret.png", img)
#画像表示
img2 = cv2.imread(path + 'ret.png')
plt.figure(figsize=(8.73/1.5, 4.66/1.5), dpi = 100)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```
