```python
import sys
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# このnotebookの存在するディレクトリのパス（必要に応じて変更！）
path = "/root/userspace/contents/materials_homework/"
gzfilename = path + 'test_img_3.png'

# 画像を読み込む。
gray = cv2.imread(gzfilename, cv2.IMREAD_GRAYSCALE)#グレースケールで読み出し
img  = cv2.imread(gzfilename)#そのままRGBで読み出し

#hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
#binary = cv2.inRange(hsv, (20, 0, 0), (50, 255, 255))
#gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)#OpenCVを使用してグレースケール化する

binary = cv2.inRange(gray, 0.01, 60)#グレースケール（2色）表示

plt.figure(figsize=(8.73/1.5, 4.66/1.5), dpi = 100)#表示する画面サイズ
plt.imshow(binary)#グレースケール化した画像を表示する
plt.show()

# 輪郭抽出
#contours = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)[0]
contours = cv2.findContours(binary,
                            cv2.RETR_EXTERNAL,
                            cv2.CHAIN_APPROX_TC89_L1)[0]

# エリア長が一定以上の部分のみ残す。
area_thresh = 40
contours = list(filter(lambda x: cv2.arcLength(x,True) > area_thresh, contours))

g_list = []#空のリストを作成
g_re_list = []#空のリストを作成
counter = 0#カウンタを定義

# 輪郭を矩形で囲む。
for cnt in contours:
    # 輪郭に外接する長方形を取得する。
    x, y, width, height = cv2.boundingRect(cnt)#一定以上の面積がある箇所の原点と縦・横長を出す
    # 描画する。
    if (width > 20) & (height > 20) & (width < 100) & (height < 100):
        #エリアの大きさが一般的な文字サイズ位である場合
        cv2.rectangle(img,#囲いを書き込む画像データ
                      (x, y),#バウンディングボックスの原点
                      (x + width, y + height),#原点からの対角点
                      color=(0, 255, 0),#色：green
                      thickness = 2)#描写線の厚み
        g_list.append( binary[y:y+height, x:x+width] )#リストに追加
        g_once = np.zeros((24,24))
        g_once = cv2.resize(binary[y:y+height, x:x+width], (24, 24))#, interpolation=cv2.INTER_NEAREST)
        g_once = np.insert(g_once, 24, 0, axis = 0)
        g_once = np.insert(g_once, 24, 0, axis = 0)
        g_once = np.insert(g_once, 24, 0, axis = 1)
        g_once = np.insert(g_once, 24, 0, axis = 1)
        g_once = np.insert(g_once, 0, 0, axis = 0)
        g_once = np.insert(g_once, 0, 0, axis = 0)
        g_once = np.insert(g_once, 0, 0, axis = 1)
        g_once = np.insert(g_once, 0, 0, axis = 1)
        g_re_list.append( g_once )
        #g_re_list.append( cv2.resize(binary[y:y+height, x:x+width], (28, 28), interpolation=cv2.INTER_NEAREST) )
        #リサイズ（28*28）してリストに追加
        counter += 1#カウンタをインクリメント

cv2.imwrite(path + "ret.png", img)
#画像表示
img2 = cv2.imread(path + 'ret.png')
plt.figure(figsize=(8.73/1.5, 4.66/1.5), dpi = 100)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import gzip
```

```python
#学習用とテスト用のデータを定義。
dl_list = [
    'emnist-balanced-train-images-idx3-ubyte.gz',
    'emnist-balanced-train-labels-idx1-ubyte.gz',
    'emnist-balanced-test-images-idx3-ubyte.gz',
    'emnist-balanced-test-labels-idx1-ubyte.gz'
]

dataset_dir = "/root/userspace/contents/materials_homework/emnist/"
```

```python
#学習用訓練データを読み出し
file_path = dataset_dir + '/' + dl_list[0]#train-images
train_data = gzip.open(file_path, 'rb')#gzファイルを開く
train_data = train_data.read()#開いたデータを読みます
train_data = np.frombuffer(train_data, np.uint8)
print(len(train_data))
```

```python
#長さが変なので加工する
train_data = np.frombuffer(train_data, np.uint8, offset=16)
print(len(train_data))

#リシェープする
train_img = train_data.reshape(-1, 28, 28)
train_img = train_img.swapaxes(1, 2)#このデータセットはX軸とY軸が入れ替わってるから変換する
print(train_img.shape)
```

```python
import tensorflow.keras as keras
from tensorflow.keras.utils import to_categorical#one-hotベクトルにするライブラリをロード

#one-hotベクトルにして学習用正解データを完成させる
train_label = to_categorical(train_data)
print(train_label.shape)
```

```python
#テスト用データを読み出し
file_path = dataset_dir + '/' + dl_list[2]#test-images
train_data = gzip.open(file_path, 'rb')#gzファイルを開く
train_data = train_data.read()#開いたデータを読みます
train_data = np.frombuffer(train_data, np.uint8)
print(len(train_data))
```

```python
#長さが変なので加工する
train_data = np.frombuffer(train_data, np.uint8, offset=16)
print(len(train_data))

test_img = train_data.reshape(-1, 28, 28)
test_img = test_img.swapaxes(1, 2)#このデータセットはX軸とY軸が入れ替わってるから変換する
print(test_img.shape)
```

```python
#テスト用データを完成させる
test_img =  test_img.reshape((test_img.shape[0], 28, 28, 1)) / 255
print(test_img.shape)
```

```python
#テスト用データを0/1のデータに加工する
test_img = np.ceil(test_img)
```

```python
#テスト用正解データを読み出す
file_path = dataset_dir + '/' + dl_list[3]#test_labelデータ
train_data = gzip.open(file_path, 'rb')
train_data = train_data.read()
train_data = np.frombuffer(train_data, np.uint8, offset = 8)
print(len(train_data))
```

```python
#one-hotベクトルにして学習用正解データを完成させる
test_label = to_categorical(train_data)
print(test_label.shape)
```

```python
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Input, Activation, add, Add, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import Sequential, Model
from sklearn.model_selection import train_test_split

from IPython.display import SVG
from tensorflow.python.keras.utils.vis_utils import model_to_dot

random_state = 42
```

```python
model = Sequential()

# 入力画像 28x28x1 (縦の画素数)x(横の画素数)x(チャンネル数)
model.add(Conv2D(16, kernel_size=(5, 5), activation='relu',#karnel_sizeがフィルタのサイズとなる
                 kernel_initializer='he_normal', input_shape=(28, 28, 1)))  # 28x28x1 -> 24x24x16
model.add(MaxPooling2D(pool_size=(2, 2)))  # 24x24x16 -> 12x12x16
model.add(Conv2D(64, kernel_size=(5, 5), activation='relu',
                 kernel_initializer='he_normal'))  # 12x12x16 -> 8x8x64
model.add(MaxPooling2D(pool_size=(2, 2)))  # 8x8x64 -> 4x4x64

model.add(Flatten())  # 4x4x64-> 1024
model.add(Dense(47, activation='softmax'))  # 1024 -> 47

model.compile(
    loss=keras.losses.categorical_crossentropy,
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
```

```python
#学習用データで訓練開始
early_stopping = EarlyStopping(monitor='val_acc', patience=15, verbose=1)#早期終了を定義
model.fit(x=train_img, y=train_label, batch_size=128, epochs=100, verbose=1,
          validation_split=0.1, callbacks=[early_stopping])#スタート
```

```python
model.save(path + 'emnist_cnn.h5')#モデルをセーブ
```

```python
#出来上がったモデルの汎化性能をチェックする
#model = load_model(path + 'mnist_cnn.h5')
test_pred = model.predict(test_img)
print(test_pred.shape)
```

```python
# 性能評価
score = model.evaluate(test_img, test_label, verbose=0)
print('Test loss:', score[0])
print('テストデータでの正解率:', score[1])
```
